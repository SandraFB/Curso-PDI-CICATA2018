# HISTOGRAMAS 
En esta práctica se aprenderá a modificar los histogramas para mejorar la calidad de la imagen. 
https://docs.opencv.org/3.1.0/d1/db7/tutorial_py_histogram_begins.html


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
     
img = cv2.imread('flor.jpg')
```

El histograma de una imagen representa la cantidad de pixeles de sus niveles de intensidad y puede generarse en escala de grises, RGB (como se muestra en la imagen de arriba) u otro modelo de color.

El eje X representa los niveles de intensidad y el eje Y representa la cantidad de pixeles.

A continuación se muestran los histogramas en las componentes Roja, Verde y Azul (RGB) por separado de una imagen.

<img src='flor.jpg' alt='flor' width='600'/>

<img src='94.png' alt='94' width='600'/>

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlabel('Niveles de intensidad (8 bits)')
    plt.ylabel('Frecuencia')
    plt.title('Histograma en RGB')
    plt.xlim([0,256])
plt.show()

```

Ahora, generemos el histograma de una región den la imagen.

```python
# Genera una máscara
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:300, 100:400] = 255
masked_img = cv2.bitwise_and(img,img,mask = mask)
```

```python
# Calcula el histograma sin la máscara
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
# Calcula el histograma con la máscara
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])
```

```python
plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
    
plt.show()
```
