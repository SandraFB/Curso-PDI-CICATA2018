# Utilizando Matplotlib
Matplotlib es una librería con una gran variedad de métodos de graficación en Python. En esta práctica se aprenderá a desplegar imágenes con esta herramienta. Se puede utilizar herramientas de zoom, guardar, etc.   

```python
import numpy as np
import cv2
```

Se importa la librería de matplotlib, un objeto pyplot y se define como plt

```python
from matplotlib import pyplot as plt
```

Se lee la imagen, se muestra...

```python
img = cv2.imread('imagen1.jpg',0)
```

plt.imshow() es la función de matplotlib para mostrar una imagen, el primer argumento es la variable que contiene la imagen, el segundo es el modelo de color, para más información visitar https://matplotlib.org

```python
plt.imshow(img, cmap = 'gray')
```

Para ocultar las marcas de las medidas se especifica, mediante la función plt.xticks() y plt.yticks() en vacío. 

```python
plt.imshow(img, cmap = 'gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
