El suavizado es una operación  simple y frecuente de procesamiento de imagen.Las razones para utilizarla son variadas, la principal es para reducir el ruido. La ejecución del suavizado conlleva aplicar un filtro a la imagen. Los filtros más comunes son lineales, en donde un pixel de salida se determina como la suma ponderada de pixeles de entrada.

h(k,l) es llamado el kernel, los coeficientes del filtro. El filtro es una ventana de coeficientes que se deslizan a través de la imagen.

Existen muchos tipos de filtros, los más comunes son los presentados a continuación:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')
```

# Convolución 2D
La función cv2.filter2D() se utiliza para convolucionar un kernel con la imagen. 
En la función de numpy np.ones((5,5),np.float32)/25 se definió el filtro de promedio siguiente:  
<img src='np.jpg' alt='np' width='200'/>


```python
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
     
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')

blur = cv2.blur(img,(5,5)) 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# SUAVIZADO GAUSSIANO
cv2.GaussianBlur().

Se especifica la altura y la anchura del kernel, el cual debe ser positivo e impar. La desviación estandar en las direccions X y Y, sigmaX y SigmaY respectivamente.  
Principalmente para eliminar ruido gaussiano. 

 blur = cv2.GaussianBlur(img,(5,5),0)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg')

blur = cv2.GaussianBlur(img,(5,5),0) 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
